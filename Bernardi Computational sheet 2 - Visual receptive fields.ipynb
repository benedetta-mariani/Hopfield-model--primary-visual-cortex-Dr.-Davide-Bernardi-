# 1) Sinusoidal grating
### s(x, y) = A cos(Kx cos Θ + Ky sin Θ − Φ)
K is the spatial frequency, Θ is the orientation, Φ is
the spatial phase and A is the contrast amplitude.

```python

## type of stimulus:
def s(x,y,A,K,theta,phi):
    return A*np.cos(K*x*np.cos(theta) +  K*y*np.sin(theta) - phi)
```

# 2) A model of receptive fields

A model of receptive fields. A receptive field (RF) describes the “sensitivity” of a neuron
to its inputs, depending on their position in the visual field. A convenient mathematical
approximation of a simple cell receptive field is provided by the Gabor function
(a) Implement a Gabor function as a model of a receptive field with preferred spatial frequency k = π
degree , preferred spatial phase ϕ = 0 and receptive field size σx = 2◦
Use the same dimension (x0, y0) as in Problem 1. Plot this Gabor receptive
field.

```python
##   simple cell receptive field is provided by the Gabor function:
## where k is the preferred spatial frequency, ϕ is the preferred spatial phase and σx, σy define
## the receptive field sizes.
def D(x,y,sigma_x, sigma_y, k, phi):
    return 1/(2*np.pi*sigma_x*sigma_y)*np.exp(-(x**2/(2*sigma_x**2) + y**2/(2*sigma_y**2)))*np.cos(k*x -phi)
```
